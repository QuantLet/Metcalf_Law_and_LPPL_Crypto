```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
from pathlib import Path
import matplotlib.pyplot as plt

res_df = pd.read_excel("\yourdata.xlsx")
res_df.head()
```

```python
ord = res_df["time"].astype("int32")
ts = [pd.Timestamp.fromordinal(d) for d in ord]
```

```python
fig, ax1 = plt.subplots(figsize=(16, 8))
ax2 = ax1.twinx()

plt.ticklabel_format(style='plain', axis = 'y') 

ax1.plot(ts, res_df['pos_conf'], color='red')
ax2.plot(ts, res_df['price'], color='black')

ax1.set_xlabel('Time', fontsize = 16)  
ax2.set_xlabel('Time', fontsize = 16) 


ax1.set_ylabel('Positive bubble indicator', fontsize = 16) 
ax2.set_ylabel('ln of Price', fontsize = 16)



plt.show()
```
